```python
titles = []
pubyear = []
pub = []
price = []
aut = []
sub = []

for dic in data:
    result_json = dic['books']
    for i in result_json:
        titles.append(i['title'])
        pubyear.append(i['publication_year'])
        price.append(i['price'])
        pub.append(i['publisher'])
        aut.append(i['authors'][0]['name'])
        sub.append(i['subtitle'])
```

```python
comp_data =  {'Title':titles, 'Author' :aut ,'Publisher':pub, 'Publish_Year': pubyear  , 'Price':price ,'Subtitles':sub}
df = pd.DataFrame(comp_data)
df
```
